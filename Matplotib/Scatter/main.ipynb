```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3]
y = [4, -6, 6]
z = [j+2 for j in y]
c = ['r', 'g', 'y']
sz = [600, 400, 300]
# color also cmap--google it

plt.scatter(x, y, linewidths=4, color=c, s=sz,
            edgecolors='g', alpha=0.5, marker='*')
plt.scatter(x, z, linewidths=4, color=c, s=sz,
            edgecolors='g', alpha=1, marker='*')

t = plt.colorbar()  # for c =[10,20,....]
t.set_label("Bar")
plt.show()
```
